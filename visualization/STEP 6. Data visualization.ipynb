```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def top_15_words(db_name, filter):
    '''Reading and sort function, which returns top most common 15 words'''
    db = pd.read_csv(db_name)
    db = db.sort_values(by=[filter],ascending=False)
    return db[1:15]
```

```python
def simple_boxplot(db, xlabel, ylabel):
    '''Simple boxplot fuction which generate plot using seaborn lib'''
    plt.figure(figsize=(15,10))

    sns.set(style="white", context="talk", font_scale=1)
    plot = sns.barplot(x=db['words'], y=db['numbers'])
    sns.despine(bottom=True)

    plt.tight_layout(h_pad=2)
    plot.set(xlabel=xlabel, ylabel=ylabel)
```

```python
simple_boxplot(react_words, 'top react words', 'numbers')
```

```python
simple_boxplot(vue_words, 'top vue words', 'numbers')
```
